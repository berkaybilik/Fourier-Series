```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_coordinates = np.linspace(-math.pi, math.pi, 628)
```

```python
def abs_sin(N, x_coordinates):
    #These calculations are based off of the Fourier Series Expansion of |sin(x)| between -pi and +pi

    y = 0
    y_coordinates = []
    for x in x_coordinates:
        for n in range(1,N):
            y = y + (4*math.cos(2*n*x))/(math.pi*(1-(4*n**2)))
        y = y + 2/(math.pi)
        y_coordinates.append(y)
        y=0
    return y_coordinates
```

```python
#Plotting with different number of terms in each sum. As the number of terms in the sum tends to infinity, the approximation made with the Fourier Series becomes exact. 
for N in [2, 5, 9, 10, 30]:
    y_coordinates = abs_sin(N, x_coordinates)
    plt.plot(x_coordinates,y_coordinates)
```

```python
def delta(N, x_coordinates):
    #These calculations are based off of the Fourier Series Expansion of the Dirac Delta function which is a function that is 0 everywhere but 0, unbounded at 0 and such that its integral over real numbers is 1.

    y_coordinates = []
    for x in x_coordinates:
        y_coordinates.append(math.sin((N+1/2)*math.pi*x)/(2*math.sin((math.pi)*x/2)))
    return y_coordinates
```

```python
x_coordinates = np.linspace(-1, 1, 200)

#Plotting for different amounts of terms in the sum. As the number of terms summed goes to infitnity, the approximation becomes exact.
for N in [2, 5, 9, 10, 30]:
    y_coordinates = delta(N, x_coordinates)
    plt.plot(x_coordinates,y_coordinates)
```
